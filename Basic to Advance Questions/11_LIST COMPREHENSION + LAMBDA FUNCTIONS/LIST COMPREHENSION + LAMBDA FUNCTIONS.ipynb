```python
nums = [1,2,3,4]

squares = list(map(lambda x: x*x, nums))

print(squares)
```

```python

nums = [1,2,3,4,5]

even = list(filter(lambda x: x % 2 == 0, nums))

print(even)
```

```python
students = [("A", 90), ("B", 70), ("C", 85)]

# sort by marks
students.sort(key=lambda x: x[1])

print(students)
```
