```python
# Multiple Return Values
def calc(a, b):
    return a + b, a - b

x, y = calc(10, 5)
print(x, y)
```

```python
# Keyword Arguments
def info(name, age):
    print(name, age)

info(age=21, name="Jitendra")
```

```python
# *args (multiple inputs)
def total(*numbers):
    print(sum(numbers))

total(1,2,3,4)
```

```python
def login(username, password):
    if username == "admin" and password == "1234":
        return "Login Successful"
    else:
        return "Invalid Credentials"

print(login("admin", "1234"))
```

# Level 2
1. Function to check even/odd
2. Function to find maximum

```python
# Function to check even/odd
def check_even_odd(x):
    if x % 2 == 0:
        print(f"{x} is Even Number")
    else:
        print(f"{x} is odd Number")

check_even_odd(5)
```

```python
# Function to find maximum
def find_max(list):
    return max(list)
    
list = [1,2,5,6,88,99,100]
find_max(list)
```
