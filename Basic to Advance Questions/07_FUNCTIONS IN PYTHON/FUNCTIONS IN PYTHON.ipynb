# Level 3 (Logic 🔥)
1. Function for factorial
2. Function for Fibonacci
3. Function for prime check

```python

# Function for factorial
def fact(x):
    result = 1
    for i in range(1,x + 1):
        result *= i
    return result
fact(5)
```

```python
# Function for factorial
def fact(x):
    result = 1
    while x>0 :
        result *= x
        x -= 1
    return result
fact(5)
```

```python
# Function for Fibonacci
def fib(x):
    list = [0,1]
    for i in range(x):
        list.append(list[-2]+ list[-1])
    return list
fib(7)
```

```python
# Function for prime check
def check_prime(x):
    is_prime = True
    for i in range(2,x):
        if x % i == 0:
            is_prime = False
            return "Not a prime number"
        else:
            return "Prime Number"
check_prime(27)
```

```python
check_prime(8)
```

```python
# Calculator using functions
def calc():
    try:

        a = int(input("Enter your number a:"))
        oprt = input("Enter your operator (+ - * /)")
        if oprt in ["+", "-", "*", "/"]:
            b = int(input("Enter your number b"))
            if oprt == "+":
                return a + b
            elif oprt == "-":
                return a - b
            elif oprt == "*":
                return a * b
            elif oprt == "/":
                return a/b if b!=0 else "Division by zero"
        else:
            return "Invalid Operator"
    except ValueError:
        return "Invalid Input: Please Enter number only"
    
calc()


```

```python
# Login system using function
def login():
    correct_username = "admin"
    correct_password = "12345"    
    
    username = input("Enter your Username:")
    password = input("Enter your Password")
    if username == correct_username and password == correct_password:
        print("Login Successfully")
    elif username != correct_username and password == correct_password:
        print("Invalid Username")
    elif password != correct_password and username == correct_username:
        print("Invaild Password")
    else:
        print("Login Unsuccessful: Please enter valid Username and Password")
    
# function call
login()

```

```python
# Menu-driven program
print("Welcome to Madhuban Foods")
print("Menu\n"
        " 1. Chai\n"
        " 2. Samosa\n"
        " 3. Biscuit\n"
        " 4. Water\n")
def orderplease():
   

    ops = {1: "Chai", 2: "Samosa", 3: "Biscuit", 4: "Water"}
    order_list = []
    while True:

        try:
            order = int(input("Enter your order number (1-4): "))
            print("You selected:", ops.get(order, "Invalid Input"))
            # int("Current Order List:", order_list)
        except ValueError:
            print("Please enter valid input") 
            continue

        choice = input("A. Add B. Delete C. That's all D. Exit ").capitalize()
        
        if choice =="A":
            if order in ops:
                order_list.append(ops[order])
                print("Added:", ops[order])
            print("Current Order List:", order_list)
        elif  choice == "B":
            if order_list:
                removed_item = order_list.pop()
                print("Removed item:", removed_item)
                print("Current Order List:", order_list)
            else:
                print("No item removed")
        elif choice == "C":
            print("Current Order List:", order_list)
            print("Final order is:", order_list)
            print("Thank You for Coming")
            break
        elif choice == "D":
            print("OK No Order\n")
            print("Thank You for Coming")
            break
        else:
            print("Invalid choice. Please select A, B, or C.")
```

```python
orderplease()
```
