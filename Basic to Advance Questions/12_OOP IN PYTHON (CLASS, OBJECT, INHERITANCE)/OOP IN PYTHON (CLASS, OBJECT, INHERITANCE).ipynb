```python
# INHERITANCE (VERY IMPORTANT 🔥)
# Parent → Child
class Animal:
    def speak(self):
        print("Animal speaks")
    
class Dog(Animal): # we inherited Aminal class into Dog class 
    def bark(self):
        print("Dog Barks")

d = Dog()  
d.speak() # Animal speaks 
d.bark() # Dog Barks
```

```python
# POLYMORPHISM

# Same function, different behavior

class Dog:
    def sound(self):
        print("Bark "* 2)

class Cat:
    def sound(self):
        print("Meow "*3)

d = Dog()
c = Cat()

d.sound()
c.sound()
```

```python
# ADVANCED UNDERSTANDING
# Instance vs Class Variable

class A:
    x = 10 # class variable


    def __init__(self):
        self.y = 20  # instance variable : variable written into inside init constructor
a = A()
print(a.x)
print(a.y)
```

```python
# Multiple Objects
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no
    
    def result(self): 
        print(f"{self.name} student roll no. {self.roll_no} scored Grade A+ in the class")

s = Student("Jitendra", 21)

s.result()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def show(self):
        print(self.balance)

acc = BankAccount(1000)
acc.deposit(500)
acc.show()
```

```python
class CAR:
    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.price = price

        # Method to display car details

    def display_details(self):
        print("Car Details")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Price: {self.price}")
            
car1 = CAR("Tesla", "Model S", 2024, 8500000)
car1.display_details()
```

```python
# Bank system

class BankAccount:
    def __init__(self, balance):
        self.balance = balance
        print(f"Current Amount Rs.{self.balance}")
    
    def withdraw(self, amount):
        self.amount = amount
        self.balance -= amount
        print(f"Withdrawl Amount Rs.{self.amount}")
    
    def add_amt(self, amount):
        self.amount = amount
        self.balance += amount
        print(f"Credited Amount Rs.{self.amount}")
    def show(self):
        print(f"Balance Amount Rs.{self.balance}")

acc = BankAccount(10000)
acc.withdraw(5000)
acc.add_amt(500)
acc.show() 
```

```python
# Employee system
class Company:
    def add_emp(self, name, ID, phone_no, working_status, address, department):
        self.name = name
        self.ID = ID
        self.phone_no = phone_no
        self.working_status = working_status 
        self.address = address 
        self.department = department
    
    def update(self):
        update = {
            "Name:": f"{self.name}",
            "ID:": f"{self.ID}",
            "Phone number:": f"{self.phone_no}",
            "Working Status:": f"{self.working_status}",
            "Address:" : f"{self.address}", 
            "Department" : f"{self.department}"

        }
        print(update)

c = Company()
c.add_emp("Jitendra", 21 ,1234567890, "YES", "Mqwer road", "AI")
c.update()
```
