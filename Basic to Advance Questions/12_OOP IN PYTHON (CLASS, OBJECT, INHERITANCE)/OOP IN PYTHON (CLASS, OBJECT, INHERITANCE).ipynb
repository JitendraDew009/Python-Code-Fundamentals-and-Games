```python
# Constructor (init)
class Student:
    def __init__(self, name, age): # self = current object
        self.name = name
        self.age = age

s1 = Student("Jitendra", 21)
print(s1)
print(s1.name)
```

```python
class Student:
    def __init__(self,name):
        self.name = name

    def greet(self):
        print(f"Hello {self.name}")

s1 = Student("Jitendra")
print(s1.name)
s1.greet()
```

```python
# ENCAPSULATION

# Data + methods ek jagah

class Bank:
    def __init__(self, balance):
        self.balance = balance
b1 = Bank(10000000000000)
b1.balance
```

```python
# INHERITANCE (VERY IMPORTANT 🔥)
# Parent → Child
class Animal:
    def speak(self):
        print("Animal speaks")
    
class Dog(Animal): # we inherited Aminal class into Dog class 
    def bark(self):
        print("Dog Barks")

d = Dog()  
d.speak() # Animal speaks 
d.bark() # Dog Barks
```

```python
# POLYMORPHISM

# Same function, different behavior

class Dog:
    def sound(self):
        print("Bark "* 2)

class Cat:
    def sound(self):
        print("Meow "*3)

d = Dog()
c = Cat()

d.sound()
c.sound()
```

```python
# ADVANCED UNDERSTANDING
# Instance vs Class Variable

class A:
    x = 10 # class variable


    def __init__(self):
        self.y = 20  # instance variable : variable written into inside init constructor
a = A()
print(a.x)
print(a.y)
```

```python
# Multiple Objects
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no
    
    def result(self): 
        print(f"{self.name} student roll no. {self.roll_no} scored Grade A+ in the class")

s = Student("Jitendra", 21)

s.result()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def show(self):
        print(self.balance)

acc = BankAccount(1000)
acc.deposit(500)
acc.show()
```
