```python
# Take 3 numbers → find largest
a = 10
b = 20
c = 90
if a > b:
    print(a)
elif b > c:
    print(b)
else:
    print(c)
```

```python
#smallest simple solution using map function
a, b, c = map(int, input("Enter 3 numbers: ").split()) # 12 45 66
print(max(a, b, c)) # 66
```

```python
# Build simple calculator (+, -, *, /)
a = float(input())
oprt = input("Choose your operator: +, -, *, /")
b = float(input())
add = a+b
sub = a-b
mul = a*b
mod = a/b
if a + b:
    print(add)
elif a - b:
    print(sub)
elif a * b:
    print(mul)
elif a / b:
    print(mod)
```

```python
# Build simple calculator (+, -, *, /)
# BETTER APPROACH
a = float(input())
oprt = input("Choose your operator: +, -, *, /")
b = float(input())
add = a+b
sub = a-b
mul = a*b
mod = a/b
if oprt == "+":
    print(add)
elif oprt == "-":
    print(sub)
elif oprt == "*":
    print(mul)
elif oprt == "/":
    print(a / b  if b!= 0 else "Error: Division by zero")
else:
    print("Invalid Operator")
# 55.2 * 2 = 110.4
# 44 / 3 = 14.666666
```
