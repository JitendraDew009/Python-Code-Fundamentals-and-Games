```python
# Basic try-except
try:
    x = 10 / 0
except:
    print("Error occurred")
```

```python
# Specific Exception
try:
    num = int(input("Enter number: "))
except ValueError:
    print("Invalid input")
```

```python
# else and finally
try:
    print("Try block")
except:
    print("Error")
else:
    print("No Error")
finally:
    print("Always runs")
```

```python
try:
    with open("data.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("File does not exist")
```

```python
# Multiple Exceptions
try:
    x = int(input())
    y = 10/x
    print(y)
except ValueError:
    print("Invalid Error")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Raise Custom Error
age = -1
if age < 0:
    raise ValueError("Age cannot be negative")
```
