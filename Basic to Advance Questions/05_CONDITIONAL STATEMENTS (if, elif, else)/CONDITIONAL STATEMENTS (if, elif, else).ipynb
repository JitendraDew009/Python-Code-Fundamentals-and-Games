```python
# Grade system (A, B, C, Fail)
marks = 56
if 90 <= marks <= 100 :
    print("Grade A")
elif 70 <= marks <= 80:
    print("Grade B")
elif 60 <= marks <= 70:
    print("Grade C")
else:
    print("Fail")

```

```python
# Check leap year
year = 2028
print("leap year" if (year % 400 ==0) or (year % 4 == 0 and year % 100 != 100) else "Not a leap year" )
```

```python
# Simple calculator using if-else
a = int(input("Enter a number: "))
oprt = input("Choose (+, - , * , /): ")
b = int(input("Enter a number: "))
if oprt == "+":
    print(a+b)
elif oprt == "-":
    print(a-b)
elif oprt == "*":
    print(a*b)
elif oprt == "/":
    print(a/b if b != 0 else "Error: Division by Zero")
else:
    print("Invalid Input")        
```
