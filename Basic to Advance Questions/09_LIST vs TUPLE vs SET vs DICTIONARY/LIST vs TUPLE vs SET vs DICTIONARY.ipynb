# Level 3 (Logic 🔥)
1. Remove duplicates using set
2. Count frequency using dictionary

```python
# Remove duplicates using set
old = ["a", "b", "c","a","a"]
unique_set = set(old)
print(unique_set)
```

```python
# Count frequency using dictionary
l = ["a", "b", "c", "a", "b", "a"]
freq = {}
for item in l:
    freq[item] = freq.get(item, 0) + 1
print(freq)
```

```python
# Build student record system
class_A = {
    "students" : {
        "name" : "Jitendra",
        "class" : 12,
        "Roll. no" : 2021,
        "subject" : {
            1 : "Maths",
            2 : "English",
            3 : "Science",
            4 : "Hindi",
            5 : "Physical Education"
        }
    } 
}
print(class_A)
```

```python
# Word frequency counter
word = "Jiiiitttteeeennddrrraaaaa"
freq = {}

for ch  in word:
    freq[ch] = freq.get(ch, 0) +1
print(freq)
```

```python
# by using collections.Counter 
from collections import Counter
word = "Jiiittenndrrrraa"
frq = Counter(word)
print(frq)
```

```python
from collections import Counter
sentence = " AI is fun and AI is rocking and trending"

word = sentence.split()

frq = Counter(word)
print(frq)
```

```python
# Find common elements between two lists
l1 = [1,2,3]
l2 = [1,3,4]
l3 = []

for i in l1:
    if i in l2 not in l3:
        l3.append(i)
print(l3)
```

```python
l1 = [1,2,3]
l2 = [1,3,4]
comman = list(set(l1) & set(l2))
print(comman)
```
