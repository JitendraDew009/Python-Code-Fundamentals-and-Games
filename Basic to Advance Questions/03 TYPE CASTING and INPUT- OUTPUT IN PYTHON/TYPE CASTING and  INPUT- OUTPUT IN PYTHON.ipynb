# Level 3 (Logic 🔥)
1. Take age and print:
 - “Minor” if <18
 - “Adult” if ≥18
2. Take price and quantity → print total bill

```python
# Take age and print:

# “Minor” if <18
# “Adult” if ≥18

age = int(input("Enter your age"))
print("Minor" if age < 18 else "Adult")
```

```python
# Take price and quantity → print total bill
item_1 = float(input("Enter your item price"))
item_2 = float(input("Enter your item price"))
item_3 = float(input("Enter your item price"))
item_4 = float(input("Enter your item price"))
total_bill = item_1 + item_2 + item_3 + item_4
print(f"Your item_1 Rs.{item_1}\n" 
      f"Your item_2 Rs.{item_2}\n"
      f"Your item_3 Rs.{item_3}\n"
      f"Your item_4 Rs.{item_4}\n"
      f"Total bill : Rs.{total_bill}")
```
