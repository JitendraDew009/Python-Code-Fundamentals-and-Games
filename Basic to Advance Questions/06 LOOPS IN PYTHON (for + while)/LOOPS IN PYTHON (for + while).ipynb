# Level 3 (Logic 🔥)
1. Prime number check
2. Factorial of number
3. Fibonacci series

```python
# Prime number check
x = 5
is_prime = True
if x<=1:
    is_prime = True
else:
    for i in range(2, x):
        if x % i == 0:
            is_prime = False
            break
if is_prime:
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
# Factorial of number using while
num = 4
fact = 1
while num > 0:
    fact = fact * num
    num = num - 1
print(fact)
```

```python
# using for loop
n = 4
result = 1
for i in range(1, n + 1):
     result *= i
print(result)
```

```python
# Fibonacci series
n = 5
fib = [0,1]
for i in range(n):
    fib.append(fib[-1]+ fib[-2])
print(fib) 
```

```python
# Menu-driven program
print("Welcome to Madhuban Foods")
print("Menu\n" 
" 1. Chai\n" 
" 2. Samosa\n" 
" 3. Biscuit\n" \
" 4. Water\n")
order_list = []

ops = {1: "Chai", 2 : "Samosa", 3 : "Biscuit", 4: "Water"}

while True:
    try:
        order = int(input("Enter your order number (1-4): "))
        print("You selected:", ops.get(order, "Invalid Input"))

    except ValueError:
        print("Please enter a valid number.")
        continue
    
    choice = input("A. Add B. Delete C. That's all ").capitalize()

    if choice == "A":
        if order in ops:
            order_list.append(ops[order])
            print("Added :",ops[order])
        else:
            print("Please enter valid input")


    elif choice == "B":
        if order_list:
            removed_item = order_list.pop()
            print("Removed:", removed_item)
            
    elif choice == "C":
        print("Thank You for Order")
        print("Your Final Order is :", order_list)
        break
    
    else:
        print("Please enter valid input")    
print("Current Order List:", order_list)
```

```python
# Menu-driven program
print("Welcome to Madhuban Foods")
print("Menu\n"
      " 1. Chai\n"
      " 2. Samosa\n"
      " 3. Biscuit\n"
      " 4. Water\n")

ops = {1: "Chai", 2: "Samosa", 3: "Biscuit", 4: "Water"}
order_list = []

while True:
    try:
        order = int(input("Enter your order number (1-4): "))
        print("You selected:", ops.get(order, "Invalid Input"))
    except ValueError:
        print("Please enter a valid number.")
        continue

    choice = input("A. Add B. Delete C. That's all: ").capitalize()

    if choice == "A":
        if order in ops:
            order_list.append(ops[order])
            print("Added:", ops[order])
        else:
            print("Invalid order number.")

    elif choice == "B":
        if order_list:
            removed_item = order_list.pop()
            print("Removed:", removed_item)
        else:
            print("No items to delete.")

    elif choice == "C":
        print("Thank you for your order!")
        print("Final Order List:", order_list)
        break

    else:
        print("Invalid choice. Please select A, B, or C.")

    print("Current Order List:", order_list)
```
