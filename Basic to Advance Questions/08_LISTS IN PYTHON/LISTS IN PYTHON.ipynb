# Level 3 (Logic 🔥)
1. Find largest number
2. Reverse list
3. Remove duplicates

```python
# find largest number
list = [1, 2, 3, 4, 5, 6]
max_list = max(list)
print(max_list)
```

```python
a = [1, 2, 3, 4, 5]
n = len(a)
max_val = a[0]
for i in range(1, n):   # 1 - 4 index only
    if a[i] > max_val:
        max_val = a[i]
print("Maximum value:", max_val)
```

```python
# Remove duplicates
list = [1, 2, 3, 0,5,8,9,66,55,147,55, 4, 5, 6]
list.sort()
unique = []
for i in list:
    if i not in unique:
        unique.append(i)
print(unique)
```

# Level 4 (0.1% 🚀)
1. Sort list without using sort()
2. Find second largest
3. Merge two lists

```python
# Sort list without using sort()
def sort(list):
    
    n = len(list)
    # this is example of bubble sort
    for i in range(n):
        for j in range(0, n - i - 1):
            if list[j] > list[j+1]:
                list[j], list[j+1] = list[j+1], list[j]
    print("Sorted list:",list) 

# Driver code
list = [1, 2, 3, 0, 5, 8, 9, 66, 55, 147, 55, 4, 5, 6]

# function calling
sort(list)          
```

```python
# Find second largest
print("Second largest number:", list[-2])
```
