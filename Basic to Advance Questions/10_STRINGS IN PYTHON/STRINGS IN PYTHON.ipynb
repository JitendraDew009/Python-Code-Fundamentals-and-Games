```python
# Reverse string
s = "Python"
s[::-1]
```

```python
# Count vowels
vowels = ["a", "e", "i", "o", "u"]
s = "Pythooon"
count = 0
for ch in s:
    if ch in  vowels:
        count += 1
print(count)
```

1. Anagram check
2. Frequency of characters
3. Clean text (like real dataset)

```python
word = "cinema".replace(" ", "").upper()
word_2 = "ice man".replace(" ", "").upper()

if sorted(word) == sorted(word_2):
    print("They are anagrams!")
else:
    print("Not anagrams.")
```

```python
def check_anagram(word, word_2):
   
    word = f"{word}".replace(" ", "").upper()
    word_2 = f"{word_2}".replace(" ", "").upper()

    if sorted(word) == sorted(word_2):
        print("They are anagrams!")
    else:
        print("Not anagrams.")

# function calling
check_anagram("cinema", "ice man")
```
